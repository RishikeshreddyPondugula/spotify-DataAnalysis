```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/spotify/spotify.csv")
```

1)What is the average danceability percentage of tracks?

```python
import matplotlib.pyplot as plt

# Calculate the average danceability percentage
average_danceability = df['danceability'].mean()

print("Average danceability percentage:", average_danceability)

# If you want to visualize the distribution of danceability using a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['danceability'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Danceability Percentage')
plt.ylabel('Frequency')
plt.title('Distribution of Danceability Percentage of Tracks')
plt.show()
```

2)What is the distribution of tracks released each year?

```python
import matplotlib.pyplot as plt

# Extract year from the 'release_date' attribute
df['release_year'] = pd.to_datetime(df['release_date']).dt.year

# Count the number of tracks released each year
tracks_per_year = df['release_year'].value_counts().sort_index()

# Plot the count of tracks released each year using a bar graph
plt.figure(figsize=(10, 6))
tracks_per_year.plot(kind='bar', color='skyblue')
plt.title('Distribution of Tracks Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

9)What are the most common genres in the dataset?


```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'spotify_data' is your DataFrame containing the Spotify data

# Extract genres from the dataset
genres = df['genres'].str.split(',').explode()

# Count the occurrences of each genre
genre_counts = genres.value_counts()

# Sort the genres based on their counts
sorted_genres = genre_counts.sort_values(ascending=False)

# Plot the distribution of genres using a pie chart or a bar graph
plt.figure(figsize=(10, 6))

# Pie chart
plt.subplot(1, 2, 1)
plt.pie(sorted_genres.values, labels=sorted_genres.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Genres (Pie Chart)')

# Bar graph
plt.subplot(1, 2, 2)
sorted_genres.plot(kind='bar', color='skyblue')
plt.title('Distribution of Genres (Bar Graph)')
plt.xlabel('Genre')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

10)What is the mode (major or minor) distribution of tracks?

```python
import matplotlib.pyplot as plt

# Count the occurrences of each mode
mode_counts = df['mode'].value_counts()

# Plot the distribution of modes using a bar graph
plt.figure(figsize=(8, 6))
mode_counts.plot(kind='bar', color='skyblue')
plt.title('Mode Distribution of Tracks')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.xticks(ticks=[0, 1], labels=['Major', 'Minor'])  # Adjust labels for major and minor modes
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

11)What is the distribution of track duration per genre?

```python
import matplotlib.pyplot as plt

# Group the dataset by genre and calculate the mean duration for each genre
genre_mean_duration = df.groupby('genres')['duration_ms'].mean()

# Plot the distribution of track duration per genre
plt.figure(figsize=(12, 6))
for genre, duration in genre_mean_duration.items():
    plt.hist(df[df['genres'] == genre]['duration_ms'], bins=20, alpha=0.7, label=genre)

plt.title('Distribution of Track Duration per Genre')
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

13)which track_id is more popular ?

```python
import matplotlib.pyplot as plt

# Find the track with the highest popularity
most_danceability_track = df.loc[df['danceability'].idxmax()]

# Extract the track_id of the most popular track
most_danceability_track_id = most_danceability_track['track_id']

# Visualize the popularity distribution
plt.figure(figsize=(8, 6))
plt.hist(df['danceability'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x=most_danceability_track['danceability'], color='red', linestyle='--', linewidth=1.5, label='Most danceability Track: ' + most_danceability_track['track_name'])
plt.title('Distribution of danceability Values')
plt.xlabel('danceability')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Display the track_id of the most popular track
print("Track ID of the most danceability track:", most_danceability_track_id)
```

14)which is album_name has highest tempo?

```python
import matplotlib.pyplot as plt

# Group the dataset by album_name and calculate the mean tempo for each album
album_tempo_mean = df.groupby('album_name')['tempo'].mean()

# Identify the album with the highest mean tempo
album_with_highest_tempo = album_tempo_mean.idxmax()

# Visualize the tempo distribution
plt.figure(figsize=(8, 6))
plt.hist(df['tempo'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x=album_tempo_mean.max(), color='red', linestyle='--', linewidth=1.5, label='Album with Highest Tempo: ' + album_with_highest_tempo)
plt.title('Distribution of Tempo Values')
plt.xlabel('Tempo')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Display the album_name with the highest tempo
print("Album with the highest tempo:", album_with_highest_tempo)
```

16)what is released_date of the tracks which has high danceability?

```python
import matplotlib.pyplot as plt

# Define threshold for high danceability
high_danceability_threshold = 0.8

# Filter the dataset to include only tracks with danceability above the threshold
high_danceability_tracks = df[df['danceability'] > high_danceability_threshold]

# Extract the release_date of these tracks
release_dates_high_danceability = high_danceability_tracks['release_date']

# Visualize the distribution of danceability
plt.figure(figsize=(8, 6))
plt.hist(df['danceability'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x=high_danceability_threshold, color='red', linestyle='--', linewidth=1.5, label='High Danceability Threshold')
plt.title('Distribution of Danceability Values')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Display the release_date of tracks with high danceability
print("Release dates of tracks with high danceability:")
print(release_dates_high_danceability)
```
